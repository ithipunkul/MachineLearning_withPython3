### List,Tuple,Set

```python
# List = [] ordered and changeable. Can Duplicate
fruits: list[str] = ["apple", "banana", "cherry","pineapple"]
print(fruits[:2])  # Step 1, start from the beginning and go up to (but not including) index 2
print(fruits[::2])  # Step 2, start from the beginning and go to the end
print(fruits[::-1])  # Step 3, start from the end and go to the beginning

# สามารถดู method ที่ใช้ได้จากด้านล่างนี้
print(dir(fruits))
print(help(fruits))
```

```python
#properties
print(len(fruits))  # Output: 4
print("melon" in fruits)  # Output: False

print(fruits.index("cherry"))  # Output: 1
print(fruits.count("apple"))  # Output: 0
```

```python
fruits[0] = "grape"
fruits.append("melon")
for i in fruits:
    print(i)
print(fruits)
```

```python
fruits.remove("banana")
fruits.insert(0, "kiwi")
fruits.sort()
print(fruits)

fruits.reverse()
print(fruits)

fruits.clear()
print(fruits)
```

### Classes

```python
class spaceship:
    def __init__(self, pilot: str, amagada: str,spaceship_name: str) -> None:  #constructor
        self.pilot = pilot
        self.amagada = amagada
        self.spaceship_name = spaceship_name
    # methods
    def fly(self) -> None:
        print(f"{self.spaceship_name} is flying!!!")
    def get_info(self, friend: str) -> None:
        print(f"{self.spaceship_name} is piloted by {self.pilot} and has {self.amagada} amagada. My friend {friend} is also on board. ><")


hailmary: spaceship = spaceship("Dr. Grace","2million","Hail Mary")
hailmary.fly()
hailmary.get_info("Rocky")
print()

Xwing: spaceship = spaceship("Luke Skywalker","1million","X-wing")
Xwing.fly()
Xwing.get_info("Leia")
```

### Dunder Method

```python
class Person:
    def __init__(self, name, age):  # Constructor
        self.name = name
        self.age = age
    
    def __str__(self) -> str:  # What prints when you do print(obj)
        return f"{self.name} is {self.age} years old"
    
    def __add__(self, other) -> str:  # What happens with +
        return Person(self.name + other.name, self.age + other.age)

Yew: Person = Person("Yew", 20)
Fang: Person = Person("Fang", 21)
Daifuku: Person = Person("Daifuku", 1)
print(Yew)  # Calls __str__ → "Yew is 20 years old"
print(Yew+ Fang)  # Calls __add__ → "YewFang is 41 years old"
print(Daifuku)
```

### Condition

```python
while True:
    user_input: str = input("You: ")
    if (user_input == "hello"):
        print("Bot: Hello there!")
    elif (user_input == "how are you?"):
        print("Bot: I'm doing well, thank you!")
    elif (user_input == "bye"):
        print("Bot: Goodbye!")
        break
    else:
        print("Bot: I don't understand.")
```

### Error handling 

```python
# use try to check for errors
a,b = 10,"fifteen"

try:
    print(a+b)
except TypeError as e:
    print(f"Please enter number in integer")
except Exception as e:
    print(f"Something went wrong: {e}")



```
