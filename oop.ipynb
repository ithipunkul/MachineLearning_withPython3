# Class

```python
class smartphone:
    def __init__(self,brand: str,model: str):  #Constructor
        self.brand = brand
        self.model = model
        self.turned_on: bool = False
    def turn_on(self) -> None:
        if self.turned_on:
            print(f"{self.brand} {self.model} is already on")
        else:
            self.turned_on = True
            print(f"{self.brand} {self.model} is now on")
    def turn_off(self) -> None:
        if self.turned_on:
            self.turned_on = False
            print(f"{self.brand} {self.model} is now off")
        else:
            print(f"{self.brand} {self.model} is already off")
    def increase_vol(self, amount: int) -> None:
        print(f"{self.brand} {self.model} volume increased by {amount}")
    def decrease_vol(self, amount: int) -> None:
        print(f"{self.brand} {self.model} volume decreased by {amount}")

Apple: smartphone = smartphone("Apple", "iPhone 17 Pro Max")
print(Apple.model)
Apple.turn_on()
Apple.turn_off()
Apple.increase_vol(10)


Samsung: smartphone = smartphone("Samsung", "Galaxy S24 Ultra")
print()
print(Samsung.model)
Samsung.decrease_vol(5)
```

# Dunder Method

```python
class Animal:
    def __init__(self, name: str, species: str, breathable: bool = False):
        self.name = name
        self.species = species
        self.breathable: bool = breathable
    def breath(self) -> None:
        if self.breathable:
            print(f"{self.name} the {self.species} is breathing")
        else:
            print(f"{self.name} the {self.species} cannot breathe")
    def make_sound(self, sound: str) -> None:
        if self.breathable:
            print(f"{self.name} the {self.species} says: {sound}")
        else:
            print(f"{self.name} the {self.species} cannot make sound because it cannot breathe")
    
    def __add__(self,other) :
        return f"{self.name} the {self.species} and {other.name} the {other.species} are friends"

    def __str__(self) -> str:  # for users, not for developers
        return f"{self.name} the {self.species}"
    def __repr__(self) -> str:  # for developers, not for users
        return f"Animal(name='{self.name}', species='{self.species}', breathable={self.breathable})" 

Duck: Animal = Animal("Donald","Duck",True)
Duck.breath()
Duck.make_sound("Quack!")
print(Duck)   # This will call the __str__ method to print the animal's name and species

print()
Fang: Animal = Animal("Fang","Not Specified")
Fang.breath()
Fang.make_sound("Grrr!")
print(Fang,"\n")   # This will call the __str__ method to print the animal's name and species        

print("Dunner method")
print(Duck + Fang)
print(repr(Duck))  # This will call the __repr__ method to print the animal's details for developers
```

# Constructor&Destructor

```python
class Employee:
    def __init__(this,name: str, salary: int,position: str):
        #public attributes
        this.name = name
        this.salary = salary
        this.position = position
        print("Employee object created")


    def __del__(this):   # Destructor method, called when the object is about to be destroyed
        print("Employee object is being deleted")
    #public method
    def showData(this) -> None:
        print(f"Name: {this.name}, Salary: {this.salary}, Position: {this.position}")

Yew: Employee = Employee("Yew", 50000, "Software Engineer")
print(Yew.name)
Yew.salary = 55000
print(Yew.salary)
Yew.showData()
del Yew
print(Employee.__class__)
```

# Encapsulation

### Access Modifier

```python
# Access modifier
# __ private คลาสลูกไม่สามารถเข้าถึงได้
# _ protected  คลาสลูกสามารถเข้าถึงได้ 
# public (no underscore)
class Employee_protected:
    def __init__(self, name: str, salary: int, position: str):
        #protected attributes
        self._name = name
        self._salary = salary
        self._position = position

    def _showData(self) -> None:
        print(f"Name: {self._name}, Salary: {self._salary}, Position: {self._position}")
Momo: Employee_protected = Employee_protected("Momo", 60000, "Data Scientist")
Momo._showData()
```

```python
class Club:
    def __init__(self, player: str, wages: int, position: str):
        self._player = player
        self.__wages = wages    # private attribute
        self._position = position
    
    def __showplanforNextMatch(self) -> None:
        print(f"4-4-2")
    def showplayer(self) -> None:
        print(f"Player: {self._player}, Position: {self._position}, Wages: {self.__wages}")

    def hackplan(self) -> None:  #getter method
        return self.__showplanforNextMatch()  # This will not work because __showplanforNextMatch is a private method
Bayern: Club = Club("Harry Kane",150000, "Striker")
Bayern.showplayer()
Bayern.wages = 1
Bayern.showplayer()
Bayern.__Club__wages = 1   
Bayern.showplayer()   #แก้ไขค่าไม่ได้ ต้องใช้ setter ตั้งค่า
Bayern.hackplan()
```

### Setter Getter

```python
class Club2:
    def __init__(self, player: str, wages: int, position: str):
        self.__player = player
        self.__wages = wages    
        self.__position = position
    
    def __showplanforNextMatch(self) -> None:
        print(f"4-3-3")
    def _showplayer(self) -> None:
        print(f"Player: {self.__player}, Position: {self.__position}, Wages: {self.__wages}")


    def setWages(self, newwages: int) -> None:  #setter method
        self.__wages = newwages

    
    def getPosition(self) -> None: # getter method field
        return self.__position
    def getPlan(self) -> None:   # getter method for private method
        return self.__showplanforNextMatch()  
manu: Club2 = Club2("Cristiano Ronaldo", 200000, "Forward")
manu.setWages(250000)
manu._showplayer()
print(manu.getPosition())
manu.getPlan()
```
