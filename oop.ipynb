# Class

```python
class smartphone:
    def __init__(self,brand: str,model: str):  #Constructor
        self.brand = brand
        self.model = model
        self.turned_on: bool = False
    def turn_on(self) -> None:
        if self.turned_on:
            print(f"{self.brand} {self.model} is already on")
        else:
            self.turned_on = True
            print(f"{self.brand} {self.model} is now on")
    def turn_off(self) -> None:
        if self.turned_on:
            self.turned_on = False
            print(f"{self.brand} {self.model} is now off")
        else:
            print(f"{self.brand} {self.model} is already off")
    def increase_vol(self, amount: int) -> None:
        print(f"{self.brand} {self.model} volume increased by {amount}")
    def decrease_vol(self, amount: int) -> None:
        print(f"{self.brand} {self.model} volume decreased by {amount}")

Apple: smartphone = smartphone("Apple", "iPhone 17 Pro Max")
print(Apple.model)
Apple.turn_on()
Apple.turn_off()
Apple.increase_vol(10)


Samsung: smartphone = smartphone("Samsung", "Galaxy S24 Ultra")
print()
print(Samsung.model)
Samsung.decrease_vol(5)
```

# Dunner Method

```python
class Animal:
    def __init__(self, name: str, species: str, breathable: bool = False):
        self.name = name
        self.species = species
        self.breathable: bool = breathable
    def breath(self) -> None:
        if self.breathable:
            print(f"{self.name} the {self.species} is breathing")
        else:
            print(f"{self.name} the {self.species} cannot breathe")
    def make_sound(self, sound: str) -> None:
        if self.breathable:
            print(f"{self.name} the {self.species} says: {sound}")
        else:
            print(f"{self.name} the {self.species} cannot make sound because it cannot breathe")
    
    def __add__(self,other) :
        return f"{self.name} the {self.species} and {other.name} the {other.species} are friends"

    def __str__(self) -> str:  # for users, not for developers
        return f"{self.name} the {self.species}"
    def __repr__(self) -> str:  # for developers, not for users
        return f"Animal(name='{self.name}', species='{self.species}', breathable={self.breathable})" 

Duck: Animal = Animal("Donald","Duck",True)
Duck.breath()
Duck.make_sound("Quack!")
print(Duck)   # This will call the __str__ method to print the animal's name and species

print()
Fang: Animal = Animal("Fang","Not Specified")
Fang.breath()
Fang.make_sound("Grrr!")
print(Fang,"\n")   # This will call the __str__ method to print the animal's name and species        

print("Dunner method")
print(Duck + Fang)
print(repr(Duck))  # This will call the __repr__ method to print the animal's details for developers
```
