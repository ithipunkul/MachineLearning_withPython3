```python
import numpy as np

vector = np.array([1, 2, 3, 4, 5] ,dtype=float)  # Need to be same data type
print(vector)
matrix = np.array([[1,2,3],[4,5,6]], dtype=int)  # must be n*m   mต้องเท่ากันทุกแถว
print(matrix)
```

```python
# Access to values
print(vector[3:])
print(vector[::2])
print(vector[2:3])
vector[0] = 999
print(vector)
```

```python
print(matrix[1,2])
print(matrix[0,1])

# 0-D array = scalar
# 1-D array = vector
# 2-D array = matrix
# N-D array = tensor (N = number > 2)
```

## Numpy Attributes

```python
arr = np.array([[1,2,3],[4,5,6]] ,dtype=np.int16)

print("Dimensions:", arr.ndim)
print("Shape:", arr.shape)
print("Size:", arr.size)
print("Data type:", arr.dtype)
```

```python
zeros = np.zeros((3,5), dtype=int)
print(zeros)
print(zeros.dtype)

ones = np.ones((3,3,3), dtype=int)
print(ones)
print(ones.dtype)
```

```python
empty = np.empty((3,5), dtype=np.int16)
print(empty)

zeros_empty = np.zeros_like(empty)
print(zeros_empty)
```

```python
print(np.arange(10))
print(np.arange(5,10))
print(np.arange(0,10,2))
```

```python
print(np.linspace(0,10,num=10, dtype=np.float16))
```
